## Image to Pencil Sketch with Python

```python
# Import libraries 

import cv2
import numpy as np
import  cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show(title, image, cmap = 'RdBu'):
    plt.imshow(image,cmap = cmap)
    plt.title(title)
```

```python
image = plt.imread("C:/Users/Lenovo/Downloads/flower.jpg")
show("Flower",image,'RdBu')
#cv2.imshow("Flower",image)
#cv2.waitKey(0)
```

## Convert to grayscale image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
show("Flower",gray_image,'gray')
```

## Invert the grayscale image

```python
inverted_image = 255 - gray_image

show("inverted_image",inverted_image,'gray')
```

## Blurr the inverted image

```python
blurred_image =cv2.GaussianBlur(inverted_image,(21,21),0)
show("blurred_image",blurred_image,'gray')
```

## Final pencil Sketch

```python
inverted_blurred = 255 - blurred_image
sketch = cv2.divide(gray_image, inverted_blurred, scale = 256.0)

show("Sketch", sketch, 'gray')
```
